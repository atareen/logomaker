```python
# Import LogoMaker
from logomaker import load_alignment, make_logo
%matplotlib inline
```

```python
! head data/PF00325_full.txt
```

```python
matrix = load_alignment(fasta_file='data/PF00325_full.txt', sequence_type='PROTEIN')
```

```python
make_logo(matrix=matrix, figsize=[10,2], occurance_threshold=.2);
```

```python
make_logo(
    matrix=matrix,
    occurance_threshold=.2,
    colors='chemistry',
    logo_type='information',
    max_stretched_character='E',
    width=.8,
    vpad=.05,
    stack_order='big_on_top',
    axes_style='classic',
    use_transparency=False,
    first_position=165,
    use_position_range=[169,192],
    figsize=[10,2],
    font_family='fantasy',
    title='CRP (Pfam PF00325).',
    save_to_file='output/wwdomain.pdf');
```
