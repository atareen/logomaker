```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.ion()

import sys
sys.path.append('../../')

import logomaker
```

```python
# Create a logo
df = pd.read_csv('../data/rnap_tau_final_all.41.matrix.txt', delim_whitespace=True)
df.head()
```

```python
#df = logomaker.validate_matrix(df)
#df = logomaker.validate_probability_mat(df)
df = logomaker.validate_information_mat(df)
df.head()
```

```python
#fig, ax = plt.subplots(figsize=[12,3])

# Create a Logo
logo = logomaker.Logo(df, center_values=True, negate_values=True)
```

```python
#fig, ax = plt.subplots(figsize=[12,3])

# Create a BaseLogo
logo = logomaker.Logo(df=df,
                      negate_values=True,
                      center_values=True,draw_now=False)

# Style glyphs
logo.style_glyphs(alpha=.9,
                  color='gray',
                  font_family='Arial Rounded MT Bold',draw_now=False)

# Add below effects
logo.style_glyphs_below(shade=.3, fade=.5,draw_now=False)

# Modify a single glyph
logo.style_single_glyph(p=30, c='A', color='dodgerblue')

# Highlight wt sequence
wt_seq = 'CAGGCTTTACACTTTATGCTTCCGGCTCGTATGTTGTGTGG'
#wt_seq1 = 'CCGG'
logo.style_glyphs_in_sequence(wt_seq, 
                              edgewidth=1.0, 
                              alpha=1.0, 
                              color='yellow')

# Draw a baseline
logo.draw_baseline(linewidth=.5)

# Style xticks
logo.style_xticks(spacing=5, anchor=25, rotation=90, fmt='%d')

# Style spines
logo.style_spines(visible=False)
logo.style_spines(visible=True, spines=('top', 'bottom'), linewidth=1)

# highlight position
logo.highlight_position(30)

# higlight positions
logo.highlight_position_range(4,9, edgecolor='b', linewidth=2, color='cyan', alpha=.3)
```

```python
import timeit

logo = logomaker.Logo(df=df,
                      negate_values=True,
                      center_values=True)

start = timeit.default_timer()

# Highlight wt sequence
wt_seq = 'CAGGCTTTACACTTTATGCTTCCGGCTCGTATGTTGTGTGG'
#wt_seq1 = 'CCGG'
logo.style_glyphs_in_sequence(wt_seq, 
                              edgewidth=1.0, 
                              alpha=1.0, 
                              color='yellow')

stop = timeit.default_timer()

print('Time: ', stop - start) 
```
